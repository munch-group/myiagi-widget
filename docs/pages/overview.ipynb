
**myiagi-widget** is a terminal trainer game ("Wax On - Wax Off" / "STEPS OF
DOOM") where you reorder the shuffled evaluation steps of a randomly
generated Python expression back into the correct order. It's built for an
introductory Python programming course, to build intuition for how Python
actually evaluates an expression -- one substitution or reduction at a time --
the same stepper engine that powers
[steps-widget](https://munch-group.org/steps-widget)'s `%%steps` cell magic
and `print-steps` command.

## Playing

```bash
myiagi
```

Digit keys select a step by its line number; arrow keys move the selected
step up/down. Get every step back into its original evaluation order to
score the round -- fewer swaps earns more points. Progress (scores, per-week
highscores, streaks) is saved to `~/.myiagi_progress.pkl`.

```
myiagi -w 3      # play as if it were course week 3
myiagi -d -2     # shift the "day" used for streak bookkeeping back two days
```


````python
from myiagi_widget.controls import leaf_prob, topic_probs
from myiagi_widget.expressions import get_expression
from steps_widget.steps import _steps

expr = get_expression(1, leaf_prob=leaf_prob, topic_probs=topic_probs)
print("Generated expression:", expr)
print()
print("Steps to reconstruct, in order:")
for i, step in enumerate(_steps(expr), start=1):
    print(f"{i:>2}  {step}")
````

## How it works

`myiagi_widget.expressions.get_expression()` builds a random `Expression`
node tree -- variables, indexing, slicing, method calls, arithmetic,
comparisons, `and`/`or` -- biased by the current course week's difficulty
ramp (`myiagi_widget.controls.topic_probs`/`leaf_prob`), then stringifies it.
That string is fed to the same `steps_widget.steps._steps()` engine
`steps-widget` uses, which produces the ground-truth ordering of
substitution/reduction steps. The game shuffles that list and challenges you
to restore it.
