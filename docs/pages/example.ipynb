A few more generated expressions and their traced steps, at whatever
difficulty `myiagi_widget.controls` computes for the current course week.

```python
from myiagi_widget.controls import course_week_nr, leaf_prob, topic_probs
from myiagi_widget.expressions import get_expression
from steps_widget.steps import _steps


def one_round():
    for _ in range(1000):
        expr = get_expression(1, leaf_prob=leaf_prob, topic_probs=topic_probs)
        try:
            steps = _steps(expr)
        except Exception:
            continue
        if 4 <= len(steps) <= 7:
            return expr, steps
    return expr, steps


print(f"Course week {course_week_nr}\n")
for round_nr in range(1, 4):
    expr, steps = one_round()
    print(f"Round {round_nr}: {expr}")
    for i, step in enumerate(steps, start=1):
        print(f"  {i:>2}  {step}")
    print()
```

## Method calls and indexing

Later weeks introduce string/list indexing, slicing, and method calls --
here forced via a hand-built `topic_probs` weighted entirely toward them, to
show the shape those rounds take.

```python
from collections import OrderedDict

later_week_topic_probs = dict(
    types=OrderedDict(dicts=0.2, lists=0.4, strings=0.4, number=0, number_literals=0),
    operations=OrderedDict(parentheses=0, len=0, abs=0, sorted=0, list=0,
                            not_op=0, logic_op=0, arithmetic_op=1.0),
)
# cumulative, like controls.py computes them
for group in later_week_topic_probs.values():
    total = 0
    for key in group:
        total += group[key]
        group[key] = total

for round_nr in range(1, 4):
    expr = get_expression(0.5, leaf_prob=0.5, topic_probs=later_week_topic_probs)
    print(f"Round {round_nr}: {expr}")
```
